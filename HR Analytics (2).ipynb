```python
# Bivariate Analysis: Attrition by Job Role
# ==========================================

# Replace the job role first
df["JobRole"] = df["JobRole"].replace(
    "Healthcare Representative Manager", "Healthcare Manager"
)

plt.figure(figsize=(10, 6))
sns.countplot(x="JobRole", hue="Attrition", data=df, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Attrition by Job Role", fontsize=14)
plt.xlabel("Job Role")
plt.ylabel("Number of Employees")
plt.legend(title="Attrition", labels=["No", "Yes"])
plt.tight_layout()
plt.show()
```

    Bivariate Analysis: Attrition by Job Role
    Data cleaning step:

    "Healthcare Representative Manager" replaced with "Healthcare Manager" for consistency.

    General observation:

    For all job roles, attrition ("Yes") is significantly higher than retention ("No").

    This suggests high turnover across the company, regardless of position.

    Highest attrition roles:

    Sales Executive shows the highest attrition count (~270 employees).

    Research Scientist also has very high attrition (~245 employees).

    Laboratory Technician follows (~200 employees).

    Moderate attrition:

    Manufacturing Director, Healthcare Manager, and Manager show mid-range attrition (~95â€“135 employees).

    Lowest attrition:

    Human Resources (~ 40 employees) and Research Director (~ 80 employees) have the least attrition.

    Insight:

    Positions with high attrition (e.g., Sales Executive, Research Scientist) might be stressful, have high performance pressure, or lower job satisfaction.

    Retention is generally weak across all roles, pointing toward possible company-wide factors (e.g., work culture, compensation, career growth).

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

    ## Correlation Heatmap

    To understand relationships between numerical features, we plot a correlation heatmap.

    - **Correlation coefficient** values range from **-1** (perfect negative) to **+1** (perfect positive).
    - Strong positive correlation means that when one variable increases, the other tends to increase as well.
    - Strong negative correlation means that when one variable increases, the other tends to decrease.

    The heatmap below highlights both strong and weak correlations in our dataset.
